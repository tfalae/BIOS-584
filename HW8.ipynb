### Q2.1. Univariate comparison (3')

<font size='4'> 
    
- Suppose that we would like to examine the utility/effect of an intervention program for patients with PTSD.
- We measure PCL5 scores at completion (`pcl5week_score.completion`) and PCL5 score at 3-month follow-up (`pcl5month_score.3_month_follow_up`). Let's assume the first score is pre-intervention and the second score is post-intervention.
- Report the summary statistics for each variable including mean, std, median, Q1, and Q3.
- Note that each patient will receive such two PCL5 scores. Use a appropriate statistical test to perform the univariate comparison. Report the outputing statistic and p-value.
- Before you run the statistic test, determine the data type and check the missingness of two columns. In particular, report the number of NA values for each variable.

```python
# Write your own code

pre_intervention = ptsd_df['pcl5week_score.completion']
post_intervention = ptsd_df['pcl5month_score.3_month_follow_up']

print("=" * 60)
print("DATA TYPE INFORMATION")
print("=" * 60)
print(f"Pre-intervention (completion) data type: {pre_intervention.dtype}")
print(f"Post-intervention (3-month follow-up) data type: {post_intervention.dtype}")
print()

print("=" * 60)
print("MISSING VALUES")
print("=" * 60)
print(f"Number of NA values in pre-intervention: {pre_intervention.isna().sum()}")
print(f"Number of NA values in post-intervention: {post_intervention.isna().sum()}")
print()

print("=" * 60)
print("SUMMARY STATISTICS: PRE-INTERVENTION (Completion)")
print("=" * 60)
pre_stats = {
    'Mean': pre_intervention.mean(),
    'Std': pre_intervention.std(),
    'Median': pre_intervention.median(),
    'Q1': pre_intervention.quantile(0.25),
    'Q3': pre_intervention.quantile(0.75)
}
for key, value in pre_stats.items():
    print(f"{key}: {value:.2f}")
print()

print("=" * 60)
print("SUMMARY STATISTICS: POST-INTERVENTION (3-Month Follow-up)")
print("=" * 60)
post_stats = {
    'Mean': post_intervention.mean(),
    'Std': post_intervention.std(),
    'Median': post_intervention.median(),
    'Q1': post_intervention.quantile(0.25),
    'Q3': post_intervention.quantile(0.75)
}
for key, value in post_stats.items():
    print(f"{key}: {value:.2f}")
print()

paired_data = pd.DataFrame({
    'pre': pre_intervention,
    'post': post_intervention
})
paired_data_clean = paired_data.dropna()

print("=" * 60)
print("PAIRED SAMPLE INFORMATION")
print("=" * 60)
print(f"Total number of patients: {len(ptsd_df)}")
print(f"Number of complete pairs (no missing values): {len(paired_data_clean)}")
print()

pre_clean = paired_data_clean['pre']
post_clean = paired_data_clean['post']

t_statistic, p_value = stats.ttest_rel(pre_clean, post_clean)

print("=" * 60)
print("PAIRED T-TEST RESULTS")
print("=" * 60)
print(f"Test statistic (t): {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print()

if p_value < 0.05:
    print("Interpretation: There is a statistically significant difference")
    print("between pre-intervention and post-intervention PCL5 scores (p < 0.05).")
else:
    print("Interpretation: There is no statistically significant difference")
    print("between pre-intervention and post-intervention PCL5 scores (p >= 0.05).")
print()

mean_diff = (pre_clean - post_clean).mean()
print(f"Mean difference (Pre - Post): {mean_diff:.2f}")
if mean_diff > 0:
    print("On average, PCL5 scores DECREASED after intervention (improvement).")
elif mean_diff < 0:
    print("On average, PCL5 scores INCREASED after intervention (worsening).")
```

### Q2.2. Multiple Linear Regression (7')

<font size='4'>

- Select columns specified in the following code chunk and create a subset dataset named `ptsd_sub_df`.
- Fit a linear regression to examine the association between `caps_intake` (continuous outcome) and the remaining covariates (as predictors) using `ptsd_sub_df`.
    - Note that all covariates ending with `_code` are categorical variables.
- Use the instruction here to write the formula for linear regression in Python.
    - https://www.statsmodels.org/stable/example_formulas.html
- Report the output page including R2, adjusted R2, and parameter estimates, SE, 95% confidence intervals, and p-values.
- Provide a brief interpretation for all significant predictors (p<0.05) excluding the intercept.
- Relevant label information includes:
    - `employment_code`: 1: Employed, 2: Unemployed, 3: Retired, 4: Disabled/Unable to work, 5: Student, 6: Other.
    - `rank_code`: 1. Enlisted, 2: Officer, 3: Other

```python
# The following column names are used for linear regression.
# Do not delete.
relevant_col_names = ['caps_intake', 'age_iop', 'gender_code', 'sexualorient_code', 'race_code', 'ethnicity_code', 
                      'education_code', 'employment_code',
                      'rank_code', 'branch_code', 'mdd_code', 'ctq_total_score', 'sexual_trauma', 'sud_code']
```

```python
# Write your own code

ptsd_sub_df = ptsd_df[relevant_col_names].copy()

print("=" * 80)
print("DATASET INFORMATION")
print("=" * 80)
print(f"Shape of subset dataset: {ptsd_sub_df.shape}")
print(f"\nMissing values:")
print(ptsd_sub_df.isna().sum())
print()

categorical_vars = [col for col in ptsd_sub_df.columns if col.endswith('_code')]
if 'sexual_trauma' in ptsd_sub_df.columns:
    categorical_vars.append('sexual_trauma')

print("=" * 80)
print("CATEGORICAL VARIABLES")
print("=" * 80)
print(categorical_vars)
print()

predictors = [col for col in ptsd_sub_df.columns if col != 'caps_intake']

formula_parts = []
for pred in predictors:
    if pred in categorical_vars:
        formula_parts.append(f'C({pred})')
    else:
        formula_parts.append(pred)

formula = 'caps_intake ~ ' + ' + '.join(formula_parts)

print("=" * 80)
print("REGRESSION FORMULA")
print("=" * 80)
print(formula)
print()

model = smf.ols(formula=formula, data=ptsd_sub_df).fit()

print("=" * 80)
print("MULTIPLE LINEAR REGRESSION RESULTS")
print("=" * 80)
print(model.summary())
print()

print("=" * 80)
print("MODEL FIT STATISTICS")
print("=" * 80)
print(f"R-squared: {model.rsquared:.4f}")
print(f"Adjusted R-squared: {model.rsquared_adj:.4f}")
print(f"F-statistic: {model.fvalue:.4f}")
print(f"Prob (F-statistic): {model.f_pvalue:.4e}")
print(f"Number of observations: {int(model.nobs)}")
print()

print("=" * 80)
print("PARAMETER ESTIMATES WITH 95% CONFIDENCE INTERVALS")
print("=" * 80)

params_df = pd.DataFrame({
    'Coefficient': model.params,
    'Std Error': model.bse,
    'CI Lower (2.5%)': model.conf_int()[0],
    'CI Upper (97.5%)': model.conf_int()[1],
    't-statistic': model.tvalues,
    'P-value': model.pvalues
})

pd.options.display.float_format = '{:.4f}'.format
print(params_df)
print()

print("=" * 80)
print("SIGNIFICANT PREDICTORS (p < 0.05)")
print("=" * 80)

significant_predictors = params_df[
    (params_df['P-value'] < 0.05) & (params_df.index != 'Intercept')
]

if len(significant_predictors) == 0:
    print("No significant predictors found at p < 0.05 level (excluding intercept).")
else:
    print(f"Number of significant predictors: {len(significant_predictors)}\n")
    print(significant_predictors)
```
